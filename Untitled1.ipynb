```python
train = pd.read_csv('temp.csv')
```

```python
channels = train.columns.values
```

```python
sum_ = {}
max_ = 0
for index, row in train.iterrows():
    count = 0
    for i in range(1, len(row)):
        count += row[i]
    sum_[index] = count
    if count > max_:
        max_ = count
ind = 0
for i in sum_.key():
    if sum_[i] == max_:
        ind = i
        
group = ''
k = ind//2
channel = []
for index, row in train.iterrows():
    if index == 10:
        group = row[0]
        for i in range(1,len(row)):
            if k >= 0 and row[i] == 1:
                channel.append(channels[i])
                k = k-1

for index, row in train.iterrows():
    if index == 10:
        for c in channel:
            train.loc[train['group_id'] == group,  c] = 0

re = reco(train)
    
rs = []
print "Getting top ",k," Recommendation for group: ", group
recos = re.recommendation_engine(group, 11)
for r in recos:
    rs.append(r[0])
    print r[0]
    
accuracy = []
for i in range(2,12):
    count = 0
    for x in rs[:i]:        
        for y in channel:
            if x == y:
                count += 1
                print x, rs[:i]
    accuracy.append(float(float(count)/float(i)))
    
import matplotlib.pyplot as plt
plt.plot(accuracy)
plt.ylabel('accuracy')
plt.xlabel('k')
plt.show()
```
